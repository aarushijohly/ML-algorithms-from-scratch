```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def findLoss(train_output, cur_y):
    ans=0
    for i in range(len(train_output)):
        ans += (train_output[i]-cur_y[i])**2
    return ans


def partialDer0(train_output, cur_y):
    ans=0
    for i in range(len(train_output)):
        ans += (cur_y[i]-train_output[i])
    return ans    

def partialDer1(train_input, train_output, cur_y):
    ans=0
    for i in range(len(train_output)):
        ans += ((cur_y[i]-train_output[i])*train_input[i])
    return ans
    
def optimizing(iterations, thetha_0, thetha_1, learning_rate, train_input, train_output):
    for i in range(iterations):
        cur_y = thetha_0 + thetha_1*train_input
        plt.plot(train_input,cur_y)
        print("total loss: ", findLoss(train_output, cur_y)/10)
        print(thetha_0, '+', thetha_1, '*x')
        partial_0 = partialDer0(train_output, cur_y)/5
        partial_1 = partialDer1(train_input, train_output, cur_y)/5
        thetha_0 = thetha_0 - learning_rate*partial_0
        thetha_1 = thetha_1 - learning_rate*partial_1
```

```python
data = {"input":[1,2,3,4,5,6], "output":[2,4,6,8,10,30]}
df = pd.DataFrame(data=data)
```

```python
fig = plt.figure(figsize = (5, 4))
plt.scatter(df['input'], df['output'])
optimizing(10, 2, 3, 0.1, df['input'], df['output'])
```
